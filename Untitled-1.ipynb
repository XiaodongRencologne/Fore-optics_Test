```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

**A cross beam scan**

```python
def plot_cross(filename):
    with open(filename,'r') as f:
        str=f.readline().split(':')[-1].split(',')
        center=[int(str[0]),int(str[1])]
    data=np.genfromtxt(filename,skip_header=1)[:,0:3]
    x=data[:,0].reshape(2,-1)[0,:]-center[0]
    y=data[:,1].reshape(2,-1)[1,:]-center[1]
    b=data[:,2].reshape(2,-1)

    fig=plt.figure(figsize=(10,4))
    plt.plot(x,10*np.log10(b[0,:]))
    plt.plot(y,10*np.log10(b[1,:]))
```

```python
plot_cross('cross/ren_80_2_160_2m_test_2.dat')
plot_cross('cross/ren_80_2_160_2m_test.dat')
```

```python
def find_focus(filename,title='',type='linear'):
    data=np.genfromtxt(filename)
    fig=plt.figure(figsize=(10,4))
    if type=='dB':
        plt.plot(data[:,0],10*np.log10(data[:,1]))
    elif type=='linear':
        plt.plot(data[:,0],data[:,1])
    plt.title(title)
```

```python
fig=plt.figure(figsize=(10,4))
for file in os.listdir('focus/'):
    if file.endswith(".dat"):
        data=np.genfromtxt(file)
        plt.plot(data[:,0],10*np.log10(data[:,1]),label=file)
plt.legend(fontsize=5)
plt.show()

```
